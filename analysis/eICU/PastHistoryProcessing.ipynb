```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
ph = pd.read_csv('pasthistory.csv')
ph
```

```python
pts= list(ph['patientunitstayid'].drop_duplicates())
```

```python
mapping = pd.read_csv('ph_mapping.csv')
mapping
```

```python
mapping = mapping[pd.notna(mapping['Paper_a'])]
pathlist = list(mapping['Pasthistorypath'])
mapping.index = mapping['Pasthistorypath']
```

```python
mappingb = mapping[pd.notna(mapping['Paper_b'])]
mappingb
```

```python
comorbidities = ['charlson', 'Acute Cerebrovascular Disease',
'Anemia',
'Atrial fibrillation',
'Blood Malignancy',
'Chronic Kidney Disease',
'Chronic obstructive pulmonary disease and bronchiectasis',
'Coronary Atherosclerosis',
'Diabetes mellitus',
'Heart valve disorders',
'Hypertension',
'Neoplasms',
'MI',
'Congestive Heart Failure',
'Healthy',
 'Unknown'                
]
```

```python
charlson = ['priormi', 'chf', 'pvd', 'diabetes', 'mld', 'sld', 'pud', 
            'cancer', 'leukemia', 'lyphoma', 'met', 'aids', 'dementia',
           'cvd', 'cpd', 'renaldisease', 'rheum', 'lymphoma']
```

```python
out = pd.DataFrame(index = pts, columns = comorbidities)
out.head()
```

```python
charlsondf = pd.DataFrame(index = pts, columns = charlson)
```

```python
counter = 0
for pt in pts[0:100]:
    counter += 1
    pdata = ph[ph['patientunitstayid'] == pt]
    
    for c in pathlist:
         if(len(pdata[pdata['pasthistorypath'] == c]) > 0):
            print(list(pdata[pdata['pasthistorypath'] == c]['pasthistorypath']))
#             out[list(mapping[mapping['Pasthistorypath'] == c]['Paper_a'])[0]][pt] = 1

#             if((pd.isna(list(mapping[mapping['Pasthistorypath'] == c]['Paper_b']))) == False):
                
#                 out[list(mapping[mapping['Pasthistorypath'] == c]['Paper_b'])[0]][pt] = 1
              
            if((pd.isna(mapping[mapping['Pasthistorypath'] == c]['Charlson'][0]) == False)):
                charlsondf[list(mapping[mapping['Pasthistorypath'] == c]['Charlson'])[0]][pt] =1
            
    if(counter%100 == 0):
        print(counter/len(pts))
```

```python
charlsondf
```

```python
counter = 0
for pt in pts[1:2]:
    counter += 1
    pdata = ph[ph['patientunitstayid'] == pt]
    out['charlson'][pt] = 0 
    
    for c in pathlist:
        if(len(pdata[pdata['pasthistorypath'] == c]) > 0):
                
            out[list(mapping[mapping['Pasthistorypath'] == c]['Paper_a'])[0]][pt] = 1

            if((pd.isna(list(mapping[mapping['Pasthistorypath'] == c]['Paper_b']))) == False):
                
                out[list(mapping[mapping['Pasthistorypath'] == c]['Paper_b'])[0]][pt] = 1
              
        
        if((pd.isna(mapping[mapping['Pasthistorypath'] == c]['score'][0]) == False)):
            out['charlson'][pt] += list(mapping[mapping['Pasthistorypath'] == c]['score'])[0]
            
    if(counter%100 == 0):
        print(counter/len(pts))
```

```python
outs = pd.DataFrame(index = pts, columns = comorbidities)
```

```python
for row in np.arange(len(ph)):   

    name = mapping[mapping['Pasthistorypath'] == ph['pasthistorypath'][row]]['Paper_a']
    
    if(len(name) > 0):
        outs[name.iloc[0]][ph['patientunitstayid'][row]] = 1
    
    if(ph['pasthistorypath'][row] in list(mappingb['Pasthistorypath'])):
        name2 = mapping[mapping['Pasthistorypath'] == ph['pasthistorypath'][row]]['Paper_b'].iloc[0]
        outs[name2][ph['patientunitstayid'][row]] = 1
    
    if(row%1000 == 0):
        print(row/len(ph))
```
